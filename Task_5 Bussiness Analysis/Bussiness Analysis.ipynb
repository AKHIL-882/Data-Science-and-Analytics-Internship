**2.3 Checking for Outliners**

```
new_data = dataset.select_dtypes(include=[np.number])
```

```

cols = new_data.columns.tolist()
for col in range(0, len(cols), 2):
  if len(cols) > col + 1:
    plt.figure(figsize=(10,4))
    plt.subplot(121)
    sns.boxplot(new_data[cols[col]])
    plt.subplot(122)
    sns.boxplot(new_data[cols[col + 1]])
    plt.tight_layout()
    plt.show()
  else:
    sns.boxplot(new_data[cols[col]])
```

**Observation**: We can find more outliers in Sales and Profit

**2.4 Removing Outliers by Winssorization**

```
from scipy.stats.mstats import winsorize 
cols = list(new_data)
for col in cols:
  if col in new_data.select_dtypes(include=np.number).columns:
    new_data[col] = winsorize(new_data[col], limits=[0.05,0.1], inclusive=(True,True))
```

**2.5. Checking whether the outliers are removed or not**

```
cols = new_data.columns.tolist()
for col in range(0, len(cols), 2):
  if len(cols) > col + 1:
    plt.figure(figsize=(10,4))
    plt.subplot(121)
    sns.boxplot(new_data[cols[col]])
    plt.subplot(122)
    sns.boxplot(new_data[cols[col + 1]])
    plt.tight_layout()
    plt.show()
  else:
    sns.boxplot(new_data[cols[col]])
```

**2.6 Finding the Frequencies of the product Subcategories**

```
plt.figure(figsize = (20,9))
sns.countplot(dataset['Sub-Category'])
plt.xticks(rotation=50, ha='right', fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Sub-Category', fontsize=20)
plt.ylabel('Frequency',fontsize=20)
```

**Observation:** We can find that **Binders, Paper, Accesssories, Phones, Furnishings** are the most top subcategories

**2.7 Finding the Profit by products**

```
high_products = dataset.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by = 'Profit', ascending = False)
sns.catplot(x = 'Sub-Category', y = 'Profit', data = high_products.head(10), kind = 'bar', aspect = 2.5)
plt.title('Top 10 Profitable Products', fontsize = 20)
plt.xticks(rotation = 50, ha = 'right', fontsize = 20)
plt.yticks(fontsize = 20)
plt.ylabel('Profit', fontsize = 20)
plt.xlabel('Products', fontsize = 20)
```

**Observation:** Here are the top most products which generates profit. Among all the products **Copiers have the highest Profit generated**.

**2.8 Frequencies of Selling Products in Cities**

```
plt.figure(figsize = (20,9))
sns.countplot(x = dataset['City'], order = (dataset['City'].value_counts().head(50)).index)
plt.xticks(rotation = 50, ha = 'right', fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('City', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20)
```

**Observation:** We can find that **New York City , Los Angeles and Philadeophia** are the top selling cities.

**2.9 Frequencies of States**

```
top_15 = dataset.groupby('State')['Quantity'].sum().reset_index().sort_values(by = 'Quantity', ascending = False)
sns.catplot(x = 'State', y = 'Quantity', data = top_15.head(15), kind = 'bar', aspect = 3)
plt.title('Top 15 States', fontsize = 15)
plt.xticks(rotation = 50, ha = 'right', fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('States', fontsize = 20)
plt.ylabel('Quantities ordered', fontsize = 20)
```

**Observation:** We can find that **California, New York and Texas** are the top states in selling products

### **3. Visualization using PieCharts**

**3.1 Finding Quantities by Region**

```
pie_region = dataset.groupby('Region')['Quantity'].sum().reset_index()
plt.figure(figsize = (15,10))
plt.pie(pie_region['Quantity'], labels = pie_region['Region'].unique(), autopct='%1.1f%%', startangle=90,
        colors = ['r', 'b', 'g', 'y'])
plt.title('Quantities ordered by each region', fontsize = 20)
```

**Observation:** We can find that most of the quantities are ordered by **West Region**

**3.2 Finding Profit by Region**

```
pie_region = dataset.groupby('Region')['Profit'].mean().reset_index()
plt.figure(figsize = (15,10))
plt.pie(pie_region['Profit'], labels = pie_region['Region'].unique(), autopct= '%.1f%%', startangle=90,
        colors = ['r', 'b', 'g', 'y'])
plt.title('Average profit divided between different regions', fontsize = 20)
```

**Observation:** We can find the most of the profit is from the West Region

**3.3 Top Selling Categories**

```
pie_categories = dataset.groupby('Category')['Quantity'].sum().reset_index()
plt.figure(figsize = (15,10))
plt.pie(x = pie_categories['Quantity'], labels = pie_categories['Category'].unique(), autopct='%1.1f%%',
        colors = ['r', 'b', 'g'])
plt.title('Products sold vs Category', fontsize = 20)
```

**Observation:** We can find that office Supplies are more Sold

**3.4 Differentiating Sub Categories Ordered**

```
plt.figure(figsize = (13,12))
plt.pie(x = dataset['Sub-Category'].value_counts(), labels = dataset['Sub-Category'].value_counts().index, autopct = '%1.1f%%',
        textprops = {'fontsize' : 15})
```

**Observation:** We can find the **Binders, Paper, Furnishings** are the top most selling subcategories.

**Conclusion**: From the above **observations and visualizations** we can conlude that Selling products in **West region** and in the cities **New York City , Los Angeles and Philadeophi**a is more preferable to generate profit
